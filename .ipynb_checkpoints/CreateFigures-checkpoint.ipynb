## **Section 4: An illustrative example: The logit-like demand function**

Solve simple logit case with very simple settings (a lot of this is default, but just in case it's altered later on, we specify it here to make sure we can replicate figures):

```python
Settings = {'I': 1000,
            'alpha': 0.5,
            'lower': 0,
            'upper': 2,
            'eps': 0,
            'sample': 'ph_gh',
            'type': 'Logit',
            'n': 200,
            'loc': 200,
            'variance_p': 1,
            'variance_g': 1,
            'corr': 0,
            'seed': 30}
Logit = WCPV.Logit(name='Logit-like demand',**Settings) # Initialize
Logit.draw_sample()
Logit.solve_npv()
Logit.solve_fp()
```

### **Hourly Utility**

```python
fig,axes = Logit.plots(['plt_uhour'],titles={'plt_uhour': None},return_fig='Yes')
fig.savefig('Figs/Logit_uhour.pdf',edgecolor='k')
```

Copy with increased $\overline{D}$:

```python
Logit_2 = WCPV.Logit(**Settings) # Initialize
Logit_2.upd_par({'lower': 0.5,
                 'upper': 3})
Logit_2.draw_sample()
Logit_2.solve_npv()
Logit_2.solve_fp()
fig,axes = Logit_2.plots(['plt_uhour'],titles={'plt_uhour': None},return_fig='Yes')
fig.savefig('Figs/Logit_uhour_v2.pdf',edgecolor='k')
```

## **Section 6: MNL**

Plot utility with same settings as in 4, except for the type = MNL

```python
Settings['type']='MNL'
Settings['BP_Ref']='level'
MNLL = WCPV.Logit(name='mnl',**Settings)
MNLL.draw_sample()
MNLL.solve_npv()
MNLL.solve_fp()
```

```python
fig,axes = MNLL.plots(['plt_uhour'],titles={'plt_uhour': None}, return_fig = 'Yes')
fig.savefig('Figs/MNLL_uhour.pdf',edgecolor='k')
```

With share bliss point:

```python
Settings['BP_Ref'] = 'share'
MNLS = WCPV.Logit(**Settings)
MNLS.draw_sample()
MNLS.solve_npv()
MNLS.solve_fp()
```

```python
fig,axes = MNLS.plots(['plt_uhour'],titles={'plt_uhour': None}, return_fig = 'Yes')
fig.savefig('Figs/MNLS_uhour.pdf',edgecolor='k')
```

Plot MNLS with a higher level of yearly $E_t$. We do not have a straightforward plotting feature to do this, but we can 'cheat' us there, by manually updating the level 'MNLS.npv['E']' which refers to the yearly level with 'no price volatility' (npv):

```python
MNLS.npv['E'] = MNLS.npv['E']*1.5
fig,axes = MNLS.plots(['plt_uhour'],titles={'plt_uhour': None}, return_fig = 'Yes')
fig.savefig('Figs/MNLS_uhour_v2.pdf',edgecolor='k')
```

Reset to obtain the right MNLS.npv['E'] again:

```python
MNLS.upd_all({}) # runs draw_sample(),solve_npv(),solve_fp() with new parameters (if any are parsed; otherwise the old values are retained)
```
